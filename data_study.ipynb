```python
import os 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
ROOT_DIR = "/home/ubuntu/ml-convection/dataset"
```

```python
t_arr_path = os.path.join(ROOT_DIR, "t_arr.npy")
ux_arr_path = os.path.join(ROOT_DIR, "ux_arr.npy")
uy_arr_path = os.path.join(ROOT_DIR, "uy_arr.npy")
```

```python
t_arr = np.load(t_arr_path)
ux_arr = np.load(ux_arr_path)
uy_arr = np.load(uy_arr_path)
```

```python
# print("p_arr.shape: ", p_arr.shape)
# print("ph_arr.shape: ", ph_arr.shape)
print("t_arr.shape: ", t_arr.shape)
print("ux_arr.shape: ", ux_arr.shape)
print("uy_arr.shape: ", uy_arr.shape)
```

```python
for i in range(0,10):
    fig, ax = plt.subplots(1, 3, figsize=(10, 10))
    
    ax[0].imshow(t_arr[i])
    ax[0].set_title(f"t_arr[{i}]")
    
    ax[1].imshow(ux_arr[i])
    ax[1].set_title(f"ux_arr[{i}]")
    
    ax[2].imshow(uy_arr[i])
    ax[2].set_title(f"uy_arr[{i}]")
    
    plt.show()
```

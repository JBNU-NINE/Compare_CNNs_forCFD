```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
true_t = np.load("/home/ubuntu/ml-convection/dataset/t_arr.npy")
print(true_t.shape)
```

```python
step = 100
end = 1000
```

```python
images = []
for i in range(0, end, step):
    images.append(np.rot90(true_t[i], k = 1))
    # print(i)
    # plt.imshow(np.rot90(true_t[i], k = 1))
    # plt.axis('off')
    # # show colorbar
    # plt.colorbar()
    # plt.show()
num_images = len(images)
print(num_images)
```

```python
from matplotlib.gridspec import GridSpec
# Define the grid size
rows = 2
cols = 5

# Calculate global vmin and vmax for consistent color scaling
vmin = min(image.min() for image in images)
vmax = max(image.max() for image in images)

# Create a figure with a GridSpec layout
fig = plt.figure(figsize=(20, 8))
# Create a GridSpec with an extra column for the colorbar
gs = GridSpec(rows, cols + 1, width_ratios=[1]*cols + [0.05], wspace=0.2, hspace=0.2)

# Loop over images and plot them
for idx, image in enumerate(images):
    row = idx // cols
    col = idx % cols
    ax = fig.add_subplot(gs[row, col])
    im = ax.imshow(image, vmin=vmin, vmax=vmax)
    ax.set_title(f'{idx+1}s')
    ax.axis('off')  # Hide axis ticks and labels

# Add the colorbar to the rightmost column
cax = fig.add_subplot(gs[:, -1])  # The last column of the GridSpec
cbar = fig.colorbar(im, cax=cax)
cbar.ax.tick_params(labelsize=12)
# cbar.set_label('Intensity', fontsize=14)

plt.show()
```
